```python
import matplotlib.pyplot as plt
import pandas as pd
file_path="railway.csv"
data_2=pd.read_csv(file_path)
eth_count=data_2["Purchase Type"].value_counts()
categories=list(eth_count.keys())
counts=list(eth_count.values)
plt.figure(figsize=(10,6))
plt.bar(categories,counts,color="skyblue")
plt.xlabel("Purchase type")
plt.ylabel("Frequency")
plt.title("plot for Purchase type")
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming data_2 is defined elsewhere and contains the purchase data
purchase_data = data_2["Purchase Type"]
total_responses = len(purchase_data)
st_percent = (purchase_data.get("Station", 0) / total_responses) * 100
on_percent = (purchase_data.get("Online", 0) / total_responses) * 100

labels = ["Station", "Online"]
sizes = [st_percent, on_percent]
colors = ["blue", "pink"]
explode = (0.1, 0)  # explode the 1st slice (Station)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Assuming data_2["Purchase Type"] is a list of purchase types
purchase_data = data_2["Purchase Type"]

# Count the occurrences of each purchase type
purchase_counts = Counter(purchase_data)

# Ensure total_responses is not zero to avoid division by zero error
total_responses = len(purchase_data)
if total_responses == 0:
    raise ValueError("No purchase data available.")

# Calculate percentages
st_count = purchase_counts.get("Station", 0)
on_count = purchase_counts.get("Online", 0)

st_percent = (st_count / total_responses) * 100
on_percent = (on_count / total_responses) * 100

labels = ["Station", "Online"]
sizes = [st_percent, on_percent]

# Verify the sizes add up to 100
if sum(sizes) != 100:
    raise ValueError("The percentages do not add up to 100.")

colors = ["blue", "pink"]
explode = (0.1, 0)  # explode the 1st slice (Station)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
